```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def analysis(train_acc,test_acc,loss,title_name,x_label,cmp):
    norm_data=[]
    for data in test_acc:
        norm=math.fabs((data-cmp)/cmp)
        norm_data.append(norm)
    print(f"The {title_name} norm:{norm_data}")
    datas=[train_acc,test_acc,loss,norm_data]
    fig,ax = plt.subplots()
    for i in range(3):
        ax.scatter(x_label,datas[i])
        ax.plot(x_label,datas[i])
    ax.scatter(x_label,datas[3],marker='x')
    ax.plot(x_label,datas[3])
    ax.legend(param_names)
    plt.title(title_name)
```

```python
# analysis
epoch_time=range(10,35,5)
train_acc=[0.907,0.920,0.943,0.948,0.954]
test_acc=[0.868,0.863,0.864,0.862,0.867]
loss=[0.202,0.219,0.155,0.136,0.128]
title_name='epoch-analysis'
param_names=['train_acc','test_acc','loss','test_rate']

analysis(train_acc,test_acc,loss,title_name,epoch_time,0.868)
```

```python
lr=[0.5,0.6,0.7,0.8,0.9,1.0]
train_acc=[0.904,0.918,0.921,0.924,0.922,0.920]
test_acc=[0.883,0.828,0.844,0.883,0.887,0.854]
loss=[0.221,0.222,0.220,0.203,0.210,0.212]
title_name='learn_rate-analysis'
param_names=['train_acc','test_acc','loss','test_rate']

analysis(train_acc,test_acc,loss,title_name,lr,0.886)
```
